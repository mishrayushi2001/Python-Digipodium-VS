```python
df['total']=df.sum(axis=1)
```

```python
df
```

```python
continent_data = df.groupby('Continent').sum()['total']
```

```python
continent_data.plot(kind='pie',autopct='%1.2f%%',figsize=(20,10))
```

```python
explode =[0,0,0.2,0,0.3,0]
continent_data.plot(kind="pie",explode=explode, shadow= True, autopct= "%1.2f%%", figsize=(20, 10), labels=None)
plt.legend(continent_data.index, loc="upper left")
```

```python
continent_data.plot(kind='pie',autopct='%1.2f%%',pctdistance= 0.7, figsize=(20,10),labels = None, wedgeprops= {'width':0.5})
plt.legend(continent_data.index, loc='upper left')
```
