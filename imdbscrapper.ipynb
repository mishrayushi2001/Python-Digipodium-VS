```python
from bs4 import BeautifulSoup
import requests
url ="https://www.imdb.com/chart/top/?ref_=nv_mv_250"
```

```python
data = requests.get(url)
data.status_code
```

```python
soup = BeautifulSoup(data.text)
```

```python
soup.find('h1').text
```

```python
rows = soup.find('tbody').findAll('tr')
```

```python
len(rows)
```

```python
for movies in rows:
    print(movies.find('td', {'class' : 'titleColumn'}).find('a').text)
    print(movies.find('td', {'class' : 'titleColumn'}).find('span').text.strip('()'))
    print(movies.find('td', {'class' : 'imdbRating'}).find('strong').text)
    
```

```python
movie_details = []

for movie in rows:
    detail = {}
    detail['name']= movies.find('td',{'class' : 'titleColumn'}).find('a').text
    detail['year']= movies.find('td',{'class' : 'titleColumn'}).find('span').text.strip('()')
    detail['ratings']= movies.find('td',{'class' : 'imdbRating'}).find('strong').text

    movie_details.append(detail)
```

```python
movie_details
```

```python
import pandas as pd
```

```python
df= pd.DataFrame(movie_details)
```

```python
df
```
